# Cpi pros/cons

### Cpi and Study before exams

```python
campus['study_before_exam'].str.replace('Studying on a regular basis', 'regular')
campus['study_before_exam'].str.replace("Pulling of all nighters :)", 'all_nighter')

print('Unique values in study column after replacements:')
print(campus['study_before_exam'].unique())
```

```python
print('Datatype before conversion:',campus.cpi.dtype)

campus['cpi'] = pd.to_numeric(campus['cpi'], errors='coerce')

campus = campus.dropna(subset=['cpi'])

campus.cpi.astype(float)
print(len(campus))
```

```python
campus = campus[(campus['cpi'] >= 4) & (campus['cpi'] <= 10)]
```

```python
campus.reset_index(drop=True, inplace=True)
```

```python
campus.cpi.mean()
```

```python
average_cpi = campus.cpi.mean()

regular_study_df = campus[(campus['study_before_exam'] == 'regular') & (campus['cpi'] > average_cpi)]
all_nighters_df = campus[(campus['study_before_exam'] == 'all_nighter') & (campus['cpi'] > average_cpi)]

# Count occurrences of each category in both DataFrames
regular_study_counts = regular_study_df.shape[0]
all_nighters_counts = all_nighters_df.shape[0]

explode=(0.1,0)

# Plot pie charts
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.pie([regular_study_counts, campus.shape[0] - regular_study_counts], labels=['Above Avg', 'Below Avg'], autopct='%1.1f%%', startangle=140, explode=explode, wedgeprops={'edgecolor':'black'})
plt.title('Regular Study')

plt.subplot(1, 2, 2)
plt.pie([all_nighters_counts, campus.shape[0] - all_nighters_counts], labels=['Above Avg', 'Below Avg'], autopct='%1.1f%%', startangle=140, explode=explode, wedgeprops={'edgecolor':'black'})
plt.title('All-nighters')

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

Interestingly most of the above average students prefer all nighters than regular study justifying the stereotype

And surprsingly most of the below average students claim to study regularly which is evident that all nighters are more efficient than regular study alone

### Trends amongst seniors and study patterns

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(data=regular_study_df, x='year', bins=5, kde=True)
plt.title('Regular Students by Year of Study')

plt.subplot(1, 2, 2)
sns.histplot(data=all_nighters_df, x='year', bins=5, kde=True, color='tab:red')
plt.title('All-nighters by Year of Study')

plt.tight_layout()
plt.show()
```

3rd yearites prefer regular studies whereas all nighters are aptly popular amongst 2nd and 4th yearites

Regular students and all nighters are almost 50% in 2nd and 3rd yearites whereas 4th yearites are almost double reliant on all nighters than regular studies
